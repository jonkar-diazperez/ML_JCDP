## API REE

```python
ree_url ="https://apidatos.ree.es"
path_renov = "/es/datos/generacion/estructura-renovables"
ree_api_param = "?start_date=2023-07-01T00:00&end_date=2025-06-30T23:59&time_trunc=month&geo_limit=ccaa&all_ccaa=allCcaa&tecno_select=1458&widget=estructura-renovables"

api_ree_url = ree_url + path_renov + ree_api_param

print(api_ree_url)

response = requests.get(api_ree_url)

if response.status_code == 200:  # Código 200 indica una respuesta exitosa.
    data = response.json()  # Analizar la respuesta JSON.
    print(data)
    #print(data["value"])
else:
    print("Error en la solicitud: ", response.status_code)
```
